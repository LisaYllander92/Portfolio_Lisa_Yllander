## #Maual testing class Shape 
Since Shapes is an Abstract Base Class (ABC), one can't instantiate it directly. An ABC is considered an incomplete blueprint because it contains abstract methods (like area and perimeter) that lack implementation.

To test the logic defined in Shapes (such as the x/y properties and the translate method), one must use a concrete subclass like Circle or Rectangle. These subclasses "carry" the inherited functionality, allowing you to test the parent class's behavior indirectly.

```python
from shapes import Shapes
try:
    s = Shapes(5,5)
except TypeError as err:
    print(err)
```

## Comparing shapes


```python
# Since they don't have the same attributes we cannot compare them directly

circle = Circle(x=0, y=0, radius=2.5)
rectangle = Rectangle(x=0, y=0, length=9, width=2)
try:
    circle == rectangle
except AttributeError:
    print(" 'Circle' object has no attribute 'length' ")
```

```python
# But, they both has an area property

print(f"The rectangle has the area: {rectangle.area} and the circle has the area: {circle.area}.")
print(f"Since they don't have the same area, the result should be 'False' = {rectangle.area == circle.area}.")
```

```python
# They also has an perimeter property

rectangle.perimeter < circle.perimeter
```
